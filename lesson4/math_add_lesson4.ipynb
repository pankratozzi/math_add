```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def equations(x):
    return np.sin(x)

x_a = np.array([])
a_vals = np.arange(0.01, 0.02, 0.0001)
x = np.linspace(100,500,100)

for i,a in enumerate(a_vals):
    x_a = np.append(x_a, float(fsolve(equations, a*x[i])))
```

```python
plt.title('$x = x(a)$')
plt.plot(a_vals, x_a)
plt.grid()
plt.show()
```
