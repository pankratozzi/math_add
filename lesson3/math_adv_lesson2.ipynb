1. Задание (в программе)

Нарисуйте график функции: $y(x) = k\cdot cos(x - a) + b$  для некоторых (2-3 различных) значений параметров k, a, b


```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)

K, A, B = [0.5, 3, 2], [1, 2, 4], [2.5, 5, -1]

plt.figure(figsize=(7,7))
for k, a, b in zip(K,A,B):
    y = k * np.cos(x - a) + b
    sign = "-" if b < 0 else "+"
    plt.plot(x, y, label=f'$y(x) = {k}\cdot cos(x - {a}) {sign} {np.abs(b)}$')
plt.grid()
plt.legend()
plt.show()
```

\begin{equation*} 
 \begin{cases}
   exp(x) + x∙(1 – y) = 1\\
   y = x^2 – 1
 \end{cases}
\end{equation*}

```python
x = np.linspace(-6, 6, 150)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x\cdot (1 – y) = 1$')

plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def equations(x):
    return np.exp(x) - 1 + x*(2-x**2)

result = set()
for a in range(-2, 5):
    result.add(int(fsolve(equations, a)[0] * 10**8)/10**8)

for res in result:
    print(f'x = {res}, y = {res**2-1}')
```
