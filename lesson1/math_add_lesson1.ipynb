```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 0.5
```

```python
def cos_to_plot(x: np.array, k: int) -> np.array:
    return np.cos(k*x)
```

```python
x = np.linspace(-15, 15, 200)
plt.figure(figsize=(12,6))
plt.plot(x, cos_to_plot(x,1), label='k=1')
plt.plot(x, cos_to_plot(x,K), label=f'k={K}')
plt.legend()
plt.grid()
plt.show()
```
